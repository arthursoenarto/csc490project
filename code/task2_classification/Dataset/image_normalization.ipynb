# This file will investigate the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/classification_dataset/training_labels.csv")
df
```

```python
df.sort_values(by=["image"])
```

```python
each_category = df[["MEL","NV","BCC","AKIEC","BKL","DF","VASC"]].sum(axis=0)
each_category
```

```python
all = each_category.sum(axis=0)
print(all)
weights = all/(each_category.to_numpy())

norm_weight = [float(i)/sum(weights) for i in weights]
norm_weight
```

```python
# Down sampling some images
mel_class = df[df["MEL"]==1]
mel_class = mel_class.iloc[:115]

nv_class = df[df["NV"]==1]
nv_class = nv_class.iloc[:115]

bcc_class = df[df["BCC"]==1]
bcc_class = bcc_class.iloc[:115]

bkl_class = df[df["BKL"]==1]
bkl_class = bkl_class.iloc[:115]

vasc_class = df[df["VASC"]==1]
vasc_class = vasc_class.iloc[:115]

akiec_class = df[df["AKIEC"]==1]
akiec_class = akiec_class.iloc[:115]

df_class = df[df["DF"]==1]

down_sampled = pd.concat([mel_class, nv_class, bcc_class, bkl_class, vasc_class, akiec_class, df_class], ignore_index=True, sort=False)
down_sampled
```

```python
down_sampled.to_csv('./data/classification_dataset/downsampled_training_labels.csv', index=False)  
```

```python
# Up sampling some images
mel_class = pd.concat([df[df["MEL"]==1], df[df["MEL"]==1]])
mel_class = mel_class.iloc[:1500]

nv_class = df[df["NV"]==1]
nv_class = nv_class.iloc[:1500]

bcc_class = pd.concat([df[df["BCC"]==1], df[df["BCC"]==1],df[df["BCC"]==1]])
bcc_class = bcc_class.iloc[:1500]

bkl_class = pd.concat([df[df["BKL"]==1], df[df["BKL"]==1]])
bkl_class = bkl_class.iloc[:1500]

vasc_class = pd.concat([df[df["VASC"]==1], df[df["VASC"]==1],df[df["VASC"]==1],df[df["VASC"]==1],df[df["VASC"]==1],df[df["VASC"]==1],df[df["VASC"]==1],df[df["VASC"]==1],df[df["VASC"]==1],df[df["VASC"]==1]])
vasc_class = pd.concat([vasc_class, vasc_class])
vasc_class = vasc_class.iloc[:1500]

akiec_class = pd.concat([df[df["AKIEC"]==1], df[df["AKIEC"]==1],df[df["AKIEC"]==1],df[df["AKIEC"]==1]])
akiec_class = pd.concat([akiec_class, akiec_class,akiec_class])
akiec_class = akiec_class.iloc[:1500]

df_class = pd.concat([df[df["DF"]==1], df[df["DF"]==1],df[df["DF"]==1],df[df["DF"]==1],df[df["DF"]==1],df[df["DF"]==1],df[df["DF"]==1],df[df["DF"]==1],df[df["DF"]==1],df[df["DF"]==1]])
df_class = pd.concat([df_class, df_class])
df_class = df_class.iloc[:1500]

up_sampled = pd.concat([mel_class, nv_class, bcc_class, bkl_class, vasc_class, akiec_class, df_class], ignore_index=True, sort=False)
up_sampled
```

```python
up_sampled.to_csv('./data/classification_dataset/upsampled_training_labels.csv', index=False)  
```
